# Matplotlib Example Workflow

```python
# 1. import matplotlib and ready it for jupyter notebook
%matplotlib inline
import matplotlib.pyplot as plt

#2. Prepare data
x = [1,2,3,4,5]
y = [0,10,12,13,14]

#3. setup plot 
fig, ax  = plt.subplots(figsize =(5,5))   #figsize specifies height and width of plot

#4. plot data
ax.plot(x,y)

#5. customize plot
ax.set(title ="Sample Plot",
      xlabel = "x-axis",
      ylabel = "y-axis")

#save and show
fig.savefig("images/sample-plot.png")
```

<h3>When to use Which method?</h3>
<ol>
    <li>to plot quickly use pyplot</li>
    <li>to plot advance plots, use OO method</li>
</ol>

```python
#plot the cholestrol level vs age>50 and color using target
over_50 = heart_disease[heart_disease.age>50]
over_50.head()
```

```python
#pyplot method
over_50.plot(kind = 'scatter',
            x = 'age',
            y= 'chol',
            c='target');
```

```python
# mix pyplot OO method
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind = 'scatter',
            x = 'age',
            y= 'chol',
            c='target',
            ax=ax);
#increasing x-axis
# ax.set_xlim([45,100])
```

```python
#oo method from scratch
fig,ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x = over_50.age,
                    y = over_50.chol,
                     c = over_50.target)
#customize plot

ax.set(title = "Age VS Chol",
      xlabel='Age',
      ylabel = 'Chol')
#add a legend
ax.legend(*scatter.legend_elements(),title='target');
#add a horizontal lines
ax.axhline(over_50.chol.mean(),linestyle='--',color='r');
```

```python
over_50.head()
```

```python
#plotting age and thalach using OO method

fig, ax = plt.subplots(figsize=(10,6))
new_scatter = ax.scatter(x = over_50.age,
                        y = over_50.thalach,
                        c = over_50.sex)
ax.set(title = "Age VS Thalach",
                        xlabel = 'Age',
                        ylabel = 'Thalach');
ax.legend(*new_scatter.legend_elements(),title='gender');

ax.axhline(over_50.thalach.mean(),linestyle='--',color='r');
```

```python
#creating figure
fig,(ax0,ax1) = plt.subplots(nrows = 2,
                            ncols = 1,
                            figsize=(10,10),
                            sharex = True)
#adding data to ax0
scatter0 = ax0.scatter(x = heart_disease.age,
                      y = heart_disease.chol,
                      c = heart_disease.target)
#styling ax0
ax0.set(title = 'Heart Disease-Comparison of AGE and CHOLESTROL Level',
       ylabel = 'CHOLESTROL')
#adding a legend
ax0.legend(*scatter0.legend_elements(),title = 'target')
#adding horizontal line
ax0.axhline(heart_disease.chol.mean(),linestyle='-');

#adding data to ax1
scatter1 = ax1.scatter(x = heart_disease.age,
                      y = heart_disease.thalach,
                      c = heart_disease.sex)
ax1.set(title = 'Heart Disease- Comparison of AGE and THALACH',
       xlabel = 'AGE',
       ylabel = 'THALACH')
#adding legend to ax1
ax1.legend(*scatter1.legend_elements(),title ="Gender")
#adding horizontal line to ax1
ax1.axhline(heart_disease.thalach.mean(),linestyle = ':');


#add a title to figure

fig.suptitle('Heart Disease Analysis',fontsize=18)
```
