```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([1,2,3,4,5,6])
y=np.array([100,200,300,400,500,600])
plt.plot(x,y);
```

```python
fig=plt.plot()
ax=fig.add_axes([1,1,1,1])
ax=fig.plot(x,y)
plt.show()
```

```python
fig,ax=plt.subplots()
ax.plot(x,x**2);
```

```python
fig,ax=plt.subplots()
ax.scatter(x,x**2)
```

```python
fig,(ax1,ax2)=plt.subplots(nrows=1,ncols=2,figsize=(10,10))
ax1.plot([1,2,3,4,5,3],[23,43,23,3,12,5])
ax2.scatter([1,2,3,4,5,3],[23,43,23,3,12,5]);
```
